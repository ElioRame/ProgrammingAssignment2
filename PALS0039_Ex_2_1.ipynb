(a) Use pandas to read the **tab-separated** file stored at this URL into a `DataFrame`:
* [https://www.phon.ucl.ac.uk/courses/pals0039/data/exercise_02/renewable_energy.tsv](https://www.phon.ucl.ac.uk/courses/pals0039/data/exercise_02/renewable_energy.tsv)

Hint: It is possible to use the Pandas [`read_csv` function](https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html) configured with the correct `delimiter` argument.

This data describes renewable energy usage as a percentage of total energy consumption between the years 1990 and 2018, broken down by country (source: [World Bank](https://data.worldbank.org/)).

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

#(a)
#ANSWER
energy_data = pd.read_csv("https://www.phon.ucl.ac.uk/courses/pals0039/data/exercise_02/renewable_energy.tsv", sep = "\t")
```

(b) Display the first 10 rows of the table:

```
#(b)
#ANSWER
energy_data.iloc[0:10]
```

(c) Set the `Country Code` column as the index of the DataFrame and remove the `Country Name` column (use this updated DataFrame in subsequent questions)

```
#(c)
#ANSWER
energy_data = energy_data.set_index("Country Code")
energy_data = energy_data.drop(columns="Country Name")
```

(d) Plot the renewable energy usage for South Africa (Country Code: `ZAF`) and the Russian Federation (Country Code: `RUS`) from 1990 to 2018 on the same axes. Include a `title`, `grid` and `legend` in the plot.

```
#(d)
#ANSWER
energy_data.loc[["ZAF", "RUS"]].T.plot(title="Renewable Energy usage", grid=True, legend=True)
plt.show()
```

(e) Display the rows in the data with the 10 highest values in 1990.

Hint: The Pandas [`sort_values` function](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sort_values.html) can be used.


```
#(e)
#ANSWER
energy_data.sort_values(by="1990", ascending=False).iloc[0:10]
```

(f) Extract data for 2018 to a `Series` and use it to display all country codes with missing data for that year. How many countries don't have data for that year?

Hint: the Series [`isna` method](https://pandas.pydata.org/docs/reference/api/pandas.Series.isna.html) can detect missing values.

```
#(f)
#ANSWER
series = energy_data["2018"]
s = sum(series.isna())
list_seriesna = list(series[series.isna()].index)

print(s, list_seriesna)
```

(g) How many countries do have data for 2018?

Hint: the Series [`notna`](https://pandas.pydata.org/docs/reference/api/pandas.Series.notna.html) and [`count`](https://pandas.pydata.org/docs/reference/api/pandas.Series.count.html) methods may be useful here.

```
#(g)
#ANSWER
series.count()
```

(h) Plot a histogram of the data for 2018.
