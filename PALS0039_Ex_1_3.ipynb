# Exercise 1.3 Matplotlib practice

In this exercise we practise using the matplotlib library for drawing graphs.

(a) Generate 100 samples using random sampling from the pre-defined distribution (in the code block below) and [plot a histogram](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html) of the values.

(b) Use numpy and the `pdf` method on the `f_pitch_distribution` object to plot the ideal distribution on the same plot as the histogram.

Hint: In order for the plots to correspond in terms of y-range the histogram will have to normalized (see the `density` parameter of matplotlib's `hist` function).

(c) Use numpy and scipy to define a new normal distribution using the statistics from the samples generated in (a) and plot the ideal shape of this new distribution

```python
import numpy as np
import scipy.stats as ss
import matplotlib.pyplot as plt
%matplotlib inline

f_pitch_mean, f_pitch_sdev = 160.0, 15.0
f_pitch_distribution = ss.norm(f_pitch_mean, f_pitch_sdev)

#(a)
#ANSWER
f_pitch_samples = f_pitch_distribution.rvs(size=100) #find sample from pre-defined distribution

#this way works as well:
counts, bins = np.histogram(f_pitch_samples, density = True)
#plt.stairs(counts, bins) #stairs = stepwise constant function as a line

plt.hist(f_pitch_samples, bins=10, range=None, density=True, histtype='bar')


#(b)
x_values = np.arange(start=f_pitch_samples.min(),
                     stop=f_pitch_samples.max(),
                     step=0.001)
plt.plot(x_values,
         f_pitch_distribution.pdf(x_values),
         label="Original distribution")

#(c)
new_distribution = ss.norm(f_pitch_samples.mean(),
                           f_pitch_samples.std())
plt.plot(x_values,
         new_distribution.pdf(x_values),
         label="New distribution")
plt.legend()
plt.show()

```
